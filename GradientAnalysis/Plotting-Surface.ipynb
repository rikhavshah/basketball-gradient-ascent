```python
import numpy as np
import matplotlib.pyplot as plt
from shot_data import shot_data
from probability_map import Processor
import matplotlib.image as mpimg
```

```python
def plot(mesh):
    all_locs = list(mesh.keys())
    locs_x, locs_y = zip(*all_locs)
    locs_x = list(set(locs_x))
    locs_y = list(set(locs_y))
    locs_x.sort()
    locs_y.sort()
    print((locs_x[0], locs_x[-1]))
    offset = 48
    locs_y = [k for k in locs_y if k <= offset]
    X, Y = np.meshgrid(locs_x, locs_y)
    # Get data to plot
    output = np.zeros((len(locs_x),len(locs_y)))
    for i in range(len(locs_x)):
        for j in range(len(locs_y)):
            output[i][j] = mesh[(locs_x[i], locs_y[j])]
    
    plt.figure(figsize = (10,10))
    plt.plot(np.linspace(-6, 8.7, 20), [-22.9 for k in range(20)], '-', color="black", linewidth=3.0)
    plt.plot(np.linspace(-6, 8.7, 20), [21 for k in range(20)], '-', color="black", linewidth=3.0)

    theta = np.linspace(-np.pi/2+.31, np.pi/2-.31, 50)
    plt.plot(23*np.cos(theta)+2, 23*np.sin(theta)-1, '-', color="black", linewidth=3.0)
    
    plt.plot(np.linspace(-6, 13, 20), [-8 for k in range(20)], '-', color="black", linewidth=3.0)
    plt.plot(np.linspace(-6, 13, 20), [7 for k in range(20)], '-', color="black", linewidth=3.0)
    plt.plot([13 for k in range(20)], np.linspace(-8,7,20), '-', color="black", linewidth=3.0)

    plt.contourf(Y, X, output.T, 20, cmap='RdGy', alpha = 0.6)
    plt.colorbar(fraction=0.041)
    plt.axes().set_aspect('equal')
    plt.xticks([])
    plt.yticks([])


    plt.savefig("all_countour.png", dpi=500, bbox_inches="tight")
    plt.show()

    return 'Done'
        
```

```python
plot(mesh)
```
